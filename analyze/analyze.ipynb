```python
from matplotlib import pyplot as plt
import numpy as np
import csv
import matplotlib.gridspec as gridspec
import matplotlib.animation as animation
from matplotlib.pyplot import MultipleLocator

#  CSV 
with open(r"ntnu_z.csv", newline='',encoding="utf-8-sig") as csvfile:
  rows = csv.reader(csvfile)
  line = []
  for row in rows:
    line.append(row)

loo=[]
looo=[]

for a in range(len(line)):  
    ll = float(line[a][0])
    loo.append(ll)
    looo.append((line[a][1]))

# 
x = []
y = []
x1 = []
y1 = []
x2 = []
y2 = []
ims = []

# 
plt.ion()
fig = plt.figure(figsize=(8, 10))
gs = gridspec.GridSpec(3,3)

# 

for i in range(len(line)):

    ax1 = plt.subplot(gs[0, :]) 
    
    plt.title("The X-axis:time  The Y-axis:For calculation", fontsize=10, color='red')
    ax1.set_xticklabels([])#x
    ax1.set_xticks([])#x
    
    
    y_major_locator = MultipleLocator(0.5)
    ax1.yaxis.set_major_locator(y_major_locator)
    
    plt.ylim(-2, 1)

    x.append((looo[i]))	# ix
    y.append((loo[i]))	# iy
    ax1.plot(x, y, color='blue', linestyle='-')  # xy

    ax1.plot(x, y, color='blue', linestyle='-')
    
    plt.pause(0.5)  # 0.0083
    plt.clf() #
```
